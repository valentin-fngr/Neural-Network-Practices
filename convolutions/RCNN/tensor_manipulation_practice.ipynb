```python
current_path = os.getcwd()
data_path = os.path.join(current_path, "data")
coco_path = os.path.join(data_path, "coco")
```

```python
def get_all_labels(): 
    
    labels = []
    label_file_path = os.path.join(coco_path, "objects-label.labels.txt") 
    print(label_file_path)
    
    with open(label_file_path, "r") as f: 
        for label in f.readlines(): 
            labels.append(label.replace("\n", ""))
    return labels
```

```python
labels = get_all_labels() 
len(labels)
```

```python
features_dict = {
    "objects/bbox" : tf.io.FixedLenSequenceFeature([], tf.float32, allow_missing=True), 
    "image": tf.io.FixedLenFeature([], tf.string), 
    "objects/area" : tf.io.FixedLenSequenceFeature([], tf.int64, allow_missing=True), 
    "image/filename" : tf.io.FixedLenFeature([], tf.string), 
    "image/id" : tf.io.FixedLenFeature([], tf.int64),
    "objects/id" : tf.io.FixedLenSequenceFeature([], tf.int64, allow_missing=True), 
    "objects/is_crowd" : tf.io.FixedLenSequenceFeature([], tf.int64, allow_missing=True),
    "objects/label" : tf.io.FixedLenSequenceFeature([], tf.int64, allow_missing=True)
}

train_path = os.path.join(coco_path, "coco-train.tfrecord-00000-of-00256") 
train_dataset = tf.data.TFRecordDataset(train_path)
```

```python
def parse_example(example): 
    parsed_example = tf.io.parse_single_example(example, features_dict) 
    return parsed_example
        
```

```python
# Training data
parsed_train_dataset = train_dataset.map(parse_example)
for sample in parsed_train_dataset: 
    decoded = cv2.imdecode(np.frombuffer(sample["image"].numpy(), np.uint8), -1)
    sample["image"] = decoded 
    print(sample)
```
