```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import sklearn.datasets
import sklearn.linear_model

def plot_decision_boundary(model, X, y):
    # Set min and max values and give it some padding
    x_min, x_max = X[0, :].min() - 1, X[0, :].max() + 1
    y_min, y_max = X[1, :].min() - 1, X[1, :].max() + 1
    h = 0.05
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    Z = model(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.ylabel('x2')
    plt.xlabel('x1')
    plt.scatter(X[0, :], X[1, :], c=y, cmap=plt.cm.Spectral)
    

def sigmoid(x):
    """
    Compute the sigmoid of x
    Arguments:
    x -- A scalar or numpy array of any size.
    Return:
    s -- sigmoid(x)
    """
    s = 1/(1+np.exp(-x))
    return s

def load_planar_dataset():
    np.random.seed(1)
    m = 400 # number of examples
    N = int(m/2) # number of points per class
    D = 2 # dimensionality
    X = np.zeros((m,D)) # data matrix where each row is a single example
    Y = np.zeros((m,1), dtype='uint8') # labels vector (0 for red, 1 for blue)
    a = 4 # maximum ray of the flower

    for j in range(2):
        ix = range(N*j,N*(j+1))
        t = np.linspace(j*3.12,(j+1)*3.12,N) + np.random.randn(N)*0.2 # theta
        r = a*np.sin(4*t) + np.random.randn(N)*0.2 # radius
        X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
        Y[ix] = j
        
    X = X.T
    Y = Y.T

    return X, Y

def load_extra_datasets():  
    N = 200
    noisy_circles = sklearn.datasets.make_circles(n_samples=N, factor=.5, noise=.3)
    noisy_moons = sklearn.datasets.make_moons(n_samples=N, noise=.2)
    blobs = sklearn.datasets.make_blobs(n_samples=N, random_state=5, n_features=2, centers=6)
    gaussian_quantiles = sklearn.datasets.make_gaussian_quantiles(mean=None, cov=0.5, n_samples=N, n_features=2, n_classes=2, shuffle=True, random_state=None)
    no_structure = np.random.rand(N, 2), np.random.rand(N, 2)
    
    return noisy_circles, noisy_moons, blobs, gaussian_quantiles, no_structure


np.random.seed(1) # set a seed so that the results are consistent
```

## Let's build a neural network with one single hidden layer

```python
X_train = X_train.T 
X_test = X_test.T

print(f"X_train shape : {X_train.shape}")
print(f"X_test shape : {X_test.shape}")
```

```python
Y_train = Y_train.T
Y_test = Y_test.T

print(f"Y_train shape : {Y_train.shape}")
print(f"Y_test shape : {Y_test.shape}")
```

```python
# activation and linear operations 

def sigmoid(Z):     
    return 1 / (1 + np.exp(-Z)) 


def relu(Z): 
    return np.maximum(Z, 0)


def compute_cost(A, Y): 
    m = A.shape[1] 
    return (- 1 / m) * np.sum(Y * np.log(A) + (1 - Y) * (np.log(1 - A)))


def linear(X, W, b): 
    assert W.shape[1] == X.shape[0] 
    return W.dot(X) + b


def relu_derivative(Z): 
    Z[Z > 0] = 1
    Z[Z <= 0] = 0
    return Z

def initialize_parameters(nb_neurones, dim, itype="random"): 
    '''
        return weights and biases initialized
        
        Args : 
            - nb_neurones : the number of neurones for each layer. ex : [4, 1]
            - dim : intuitively, the dimension of one activation ex : [100, 4]
            
        Output: 
            parameters : a dict of weights and biases for each layers
    '''
    assert len(nb_neurones) == len(dim)
    assert nb_neurones[0] == dim[1]
    
    d = {}
    print(f"Intitializing for {len(nb_neurones)} layers using the {itype} method ... ") 
    if itype == "zeros": 
        
        for i in range(len(nb_neurones)): 
            print(f"Init for layer : {i}") 
            W = np.zeros(([nb_neurones[i], dim[i]]))
            b = np.zeros((nb_neurones[i], 1)) 

            # save 
            d[f"W{i+1}"] = W 
            d[f"b{i+1}"] = b
            
    elif itype == "random": 
        for i in range(len(nb_neurones)): 
            print(f"Init for layer : {i}") 
            W = np.random.randn(nb_neurones[i], dim[i]) * 0.01
            b = np.random.randn(nb_neurones[i], 1) * 0.01 

            # save 
            d[f"W{i+1}"] = W 
            d[f"b{i+1}"] = b
            
    elif itype == "he": 
        for i in range(1, len(nb_neurones) + 1): 
            print(f"Init for layer : {i}") 
            W = np.random.randn(nb_neurones[i-1], dim[i-1]) * np.sqrt(2/nb_neurones[i-1])
            b = np.random.randn(nb_neurones[i-1], 1) * 0.01 
            d[f"W{i}"] = W 
            d[f"b{i}"] = b
            

            
        
    print(f"Successfully initialized parameters for {len(nb_neurones)} layers ! \n")
    
    return d
```

```python
# check for initialize_parameters
d = initialize_parameters([4, 1], [2, 4])

for k, v in d.items(): 
    print(f"{k} : {v.shape}")
```

```python
# neural networks functions 

def forward_pass(X, parameters, Y): 
    # linear => relu => linear => sigmoid => cost 

    W1, b1, W2, b2 = parameters["W1"], parameters["b1"], parameters["W2"], parameters["b2"]
    
    Z1 = linear(X, W1, b1)
    A1 = relu(Z1) 
    Z2 = linear(A1, W2, b2) 
    A2 = sigmoid(Z2)
    
    cost = compute_cost(A2, Y)
    
    cache = {
        "Z1" : Z1, 
        "A1" : A1, 
        "Z2" : Z2, 
        "A2" : A2
    }
    
    return cost, cache
```

```python
def backward_pass(X, cache, parameters, Y): 
    
    Z1, A1, Z2, A2 = cache["Z1"], cache["A1"], cache["Z2"], cache["A2"]
    
    
    m = Z1.shape[1]
    
    # compute grads  
    # second layer    
    dZ2 = A2 - Y
    dW2 = np.dot(dZ2, A1.T) *(1/m) 
    db2 = np.sum(dZ2, axis=1, keepdims=True) * (1/m)
    

    # first layer
    dZ1 = np.multiply(parameters["W2"].T.dot(dZ2),  relu_derivative(Z1))
    dW1 = dZ1.dot(X.T) * (1/m)
    db1 = np.sum(dZ1, axis=1, keepdims=True) * (1/m)
    grads = {
        "dW1" : dW1, 
        "db1" : db1, 
        "dW2": dW2, 
        "db2" : db2
    }

    return grads


def update_parameters(parameters, grads, learning_rate): 
    assert len(parameters.keys()) == len(grads.keys())
    
#     for param in parameters: 
#         parameters[param] = parameters[param] - learning_rate * grads[f"d{param}"]

    W1 = parameters["W1"]
    W2 = parameters["W2"] 
    b1 = parameters["b1"] 
    b2 = parameters["b2"] 
    
    dW1 = grads["dW1"] 
    dW2 = grads["dW2"] 
    db1 = grads["db1"] 
    db2 = grads["db2"]
    
    assert dW1.shape == W1.shape 
    assert dW2.shape == W2.shape 
    assert db1.shape == db1.shape 
    assert db2.shape == db2.shape
        
    W1 = W1 - learning_rate * dW1
    W2 = W2 - learning_rate * dW2
    b1 = b1 - learning_rate * db1
    b2 = b2 - learning_rate * db2
    
    parameters["W1"] = W1
    parameters["W2"] = W2 
    parameters["b1"] = b1
    parameters["b2"] = b2
    
    return parameters
```

```python
# predict 

def predict(Y_true, A): 
    print(Y_true.shape)
    print(A.shape)
    assert Y_true.shape == A.shape
    
    predictions = np.zeros((1, A.shape[1]))
    predictions[A > 0.5] = 1
    predictions[A <= 0.5] = 0 
    print(f"Obtained : {(Y_true == predictions).mean() * 100} % ")
    
    return predictions
```

```python
# entire model 

def neural_network(X_train, Y_train, nb_neurones, dims, learning_rate=0.01, epochs=100, ax=None, show_print=False, itype="random"): 
    
    # init
    parameters = initialize_parameters(nb_neurones, dims, itype)
    cost_history = []
    for i in range(epochs): 
    
        # forward pass 
        cost, cache = forward_pass(X_train, parameters, Y_train)
        if i % 100 == 0: 
            cost_history.append(cost)
            if show_print: 
                print(f"epoch number : {i} -----------> cost function : {cost}")
            
        #backwards prop 
        grads = backward_pass(X_train, cache, parameters, Y_train)
        # update 
        parameters = update_parameters(parameters, grads, learning_rate)
    
    # predicting 
    _, cache = forward_pass(X_train, parameters, Y_train) 
    
            
    print(f"Training accuracy for model with : nb_neurones : {nb_neurones} | learning_rate : {learning_rate}, epochs : {epochs} :" )
    predict(Y_train, cache["A2"])
    print("\n")
    if ax: 
        plt.plot(range(epochs//100), cost_history, axis=ax)
    else: 
        plt.plot(range(epochs//100), cost_history)
    return parameters
```

```python
parameters = neural_network(X_train, Y_train, [4, 1], [X_train.shape[0], 4], 0.1, 50000)
print("Testing accuracy : ")
_, cache = forward_pass(X_test, parameters, Y_test) 
predict(Y_test, cache["A2"])
```

```python
def plot_models(X_train, Y_train, nb_neurones, dims, learning_rate, epochs):   
    '''
        Plot multiple models
        
        X_train : X train 
        Y_train : Y train vector
        nb_neurones : list of list containing the number of neurones for each model 
                        ex : [ [4, 1], [10, 1]]
        dim: list of list of dimensions for each model 
                        ex : [ [100, 4], [100, 3]]
        learning_rate : list of learning_rates for each model : ex : [0.01, 0.1, 0.001] 
        epochs : list of epochs for each model, ex : [1000, 2000, 10000]
    '''
    titles = []
    
    for i in range(len(nb_neurones)): 
        for lr in learning_rates:
            for epoch in epochs: 
                print(f"Training with : {nb_neurones[i]} | {dims[i]} | {lr} | {epoch}")
                parameters = neural_network(X_train, Y_train, nb_neurones[i], dims[i], lr, epoch, show_print=False)
                titles.append([f"{lr} {epoch}  {nb_neurones[i][0]}"])
                
    plt.legend(titles)
    plt.ylabel("Cost") 
    plt.xlabel("epoch in ")
    plt.figure(figsize=(14, 14))
    return plt.show()
```

```python
nb_neurones = [
    [4, 1], 
    [20, 1], 
    [1, 1]
] 

dims = [
    [X_train.shape[0], 4], 
    [X_train.shape[0], 20], 
    [X_train.shape[0], 1]
]

learning_rates = [1, 0.1, 0.01]
epochs = [10000, 30000] 

plot_models(X_train, Y_train, nb_neurones, dims, learning_rates, epochs) 
```
